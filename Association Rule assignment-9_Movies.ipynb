1. Association rules with 10% Support and 70% confidence

```python
# with 10% support
frequent_itemsets=apriori(movie2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

observation:-
1)The most popular movies in the dataset (with support 60-70%) are Sixth Sense, Gladiator, and Patriot. These movies have a significant presence in the transactions.

2)Movies like LOTR1, Harry Potter1, LOTR2, Harry Potter2, LOTR, Braveheart, and Green Mile have a lower support, indicating they are less frequently bought compared to the popular ones.

3)The combination of Gladiator and Sixth Sense is the most frequent (10%). This suggests that a considerable portion of customers who buy Gladiator also buys Sixth Sense.

4)Patterns like (Gladiator, Patriot, Sixth Sense) and (LOTR, Gladiator, Green Mile) indicate a group of customers who prefer action and thriller genres together.

5)There are niche preferences indicated by rules like (LOTR, Green Mile, Sixth Sense) and (LOTR1, Green Mile, Sixth Sense, Harry Potter1). These suggest a smaller group of customers who have specific taste in movies, possibly a mix of fantasy and drama.

6)Certain combinations like (LOTR, Gladiator, Green Mile) suggest a cross-genre interest. It indicates that some customers enjoy movies from different genres.

7)There are complex patterns like (LOTR2, LOTR1, Green Mile, Harry Potter1) and (LOTR1, Green Mile, Sixth Sense, LOTR2, Harry Potter1) indicating a very specific and complex taste in movies. This kind of insight could be valuable for targeted marketing to this specific group of customers.

```python
# 70% confidence
rules_1=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules_1
```

1)The Associations rule between "Sixth Sense" and "Gladiator" and vice versa remain strong with a confidence of 83.33% and 71.43% respectively, even at the higher confidence threshold. This indicates a robust relationship between these movies.

2)"LOTR1" and "Sixth Sense" still have a moderate confidence of 50%. Despite the increase in confidence threshold, this association remains stable.

3)Many of the association rules still have lift values greater than 1, indicating that the antecedent and consequent items are more likely to be bought together than individually. For instance, the association between "(LOTR1) and (LOTR2, Green Mile, Sixth Sense, Harry Potter1)" has a lift of 5. This suggests a strong relationship between LOTR1 and this combination of movies.

4)There are still no strong negative associations, reinforcing that there are no combinations of movies that customers actively avoid buying together at this confidence leve

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules_1[rules_1.lift>1]
```

1)The movies "Gladiator" and "Sixth Sense" are popular on their own, and strong with a lift value greater than 1 and occurring in at least 10% of the transactions.
  Other movies like "Patriot" and "LOTR2" are also frequently bought together than individually, meeting the lift value greater than 1 and  10% support threshold.
  There might be a trend where certain pairs of movies are preferred together by customers.

2)There is a perfect association between "LOTR" and "Sixth Sense" with a lift of 1.67, suggesting that whenever "LOTR" is bought, "Sixth Sense" is also bought, and vice versa.

4)These insights can be valuable for marketing strategies. For instance, if "Gladiator" and "Patriot" are frequently bought together, a marketing campaign promoting these two movies as a bundle deal could be effective.
Movie recommendations for online platforms: If a customer is viewing "Sixth Sense", recommendations can be made for "Gladiator" or "Patriot" based on the frequent itemsets.

```python
import matplotlib.pyplot as plt

# Assuming rules_df is your DataFrame containing association rules
plt.figure(figsize=(10, 6))
plt.bar(rules_1.index, rules_1['support'], label='Support')
plt.plot(rules_1.index, rules_1['confidence'], marker='o', label='Confidence')
plt.plot(rules_1.index, rules_1['lift'], marker='o', label='Lift')
plt.xlabel('Association Rules')
plt.ylabel('Metrics')
plt.legend()
plt.show()
```

2. Association rules with 5% Support and 90% confidence

```python
# with 5% support
frequent_itemsets2=apriori(movie2,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

1)Movies "Gladiator" and "Sixth Sense" continue to be popular, appearing in at least 5% of the transactions.
  "Patriot" is also a popular single movie, with a support of 0.6.
    
2)The pairs "Sixth Sense", "Gladiator" and "Gladiator", "Patriot" are still frequent, with supports of 0.5 and 0.6 respectively. This indicates that these movies are often bought together.
  Other pairwise associations like "Sixth Sense", "Patriot" and "Gladiator", "Green Mile" also have support greater than 5%.
    
3)There are several complex associations involving three or more movies, such as "Sixth Sense", "Gladiator", "Patriot" and "Sixth Sense", "Gladiator", "Patriot", "Green Mile", "LOTR1". These complex associations suggest certain groups of movies that are frequently bought together.

4)The presence of numerous moviesets with various combinations indicates that customers have diverse movie preferences.
  Even with a lower support threshold, there are still diverse patterns, suggesting a broad range of customer preferences.
    
5)These associations provide insights into customer behavior, indicating which movies are often considered together. This information is valuable for understanding customer preferences and tailoring services accordingly.

```python
# 90% confidence
rules3=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules3
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules3[rules3.lift>1]
```

1)The associations between "Sixth Sense" and "Gladiator" and "Patriot" and "Sixth Sense" remain strong even at a 90% confidence level, indicating a high likelihood of these movies being bought together.

2)Similar to before, there is a perfect association between "LOTR" and "Sixth Sense" with a lift of 1.67. This means that whenever "LOTR" is bought, "Sixth Sense" is also bought, and vice versa.

3)Many other associations continue to have high lift values, such as the association between "(LOTR1) and (LOTR2, Green Mile, Sixth Sense, Harry Potter1)" with a lift of 5. This suggests a strong relationship between these items.

4)With the high confidence and lift values can indicate the customers are likely to be satisfied with these combinations of movies and it can be leading increasing of customers positive reviews.

5)The Businesses can confidently focus on marketing and sales efforts on the   movies involved in these strong associations.
  Bundling these movies together in promotions or offering discounts           on combinations could be a successful strategy.
